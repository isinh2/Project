## Which month has the highest Sells


```python
data_all['Order Month'] = data_all['Order Month'].astype('int64')
```

```python
data_all.info()
```

```python
data_all['Order Month'] = data_all['Order Month'].astype(int)
```

```python
data_all['Price Each'] = data_all['Price Each'].astype(float)
```

```python
data_all.head()
```

```python
data_all['Quantity Ordered'] = data_all['Quantity Ordered'].astype(int)
```

```python
data_all.info()
```

```python
data_all['Total Bill'] = data_all['Quantity Ordered'] * data_all['Price Each']
```

```python
data_all.head()
```

```python
data_all_month = data_all.groupby('Order Month')[['Quantity Ordered','Price Each','Total Bill']].sum()
```

```python
data_all_month.head()
```

```python
data_all_month = data_all_month.reset_index()
```

```python
data_all_month = data_all_month.sort_values('Total Bill',ascending=False)
```

```python
data_all_month.head()
```

```python
sns.set_style('darkgrid')
```

```python
sns.barplot(x='Order Month',y='Total Bill', data=data_all_month)
plt.plot(data_all_month['Order Month'],data_all_month['Price Each'] , 'o-r')
plt.plot(data_all_month['Order Month'],data_all_month['Quantity Ordered'],'x--b');
```

## What City has highest number Of Sell

```python
plt.figure(figsize=(8,5))
sns.barplot(x='city',y='Total Bill',data=data_all_city)
plt.xticks(rotation = 45);
plt.plot(data_all_city['city'],data_all_city['Price Each'], 'x--r')
plt.plot(data_all_city['city'],data_all_city['Quantity Ordered'], 'o-b');
```
